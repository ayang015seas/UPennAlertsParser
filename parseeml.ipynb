```python
import datetime
import json
import eml_parser
import os
import csv


def json_serial(obj):
  if isinstance(obj, datetime.datetime):
      serial = obj.isoformat()
      return serial


with open('UPennAlert issued at 1_18_20 3_21 AM.eml', 'rb') as fhdl:
  raw_email = fhdl.read()

ep = eml_parser.EmlParser()
parsed_eml = ep.decode_email_bytes(raw_email)



allemails = []

def readFile(filename):
    file1 = open(filename, 'r')
    Lines = file1.readlines()
    date = 0
    subj = None
    text = None
    
    count = 0
    # Strips the newline character
    for line in Lines:
        if (count == 2):
            date = line.strip()
        splitline = line.split(" ")
        if (splitline[0] == "Subject:"):
            subj = line.strip()
        if (splitline[0] == "Alexander"):
            text = Lines[count + 1]
        count += 1

#         print("Line{}: {}".format(count, line.strip()))
    
    thisdict = {
      "date": date,
      "subject": subj,
      "msg": text
    }
    file1.close()
    return thisdict

def getAllFiles():
    all_emails = []
    all_files = os.listdir(os.getcwd())
    for i in range(0, len(all_files)):
        file_name_split = all_files[i].split(" ")
        if (file_name_split[0] == "UPennAlert"):
            ret_value = readFile(all_files[i])
            all_emails.append(ret_value)
    return all_emails



csv_columns = ['date','subject','msg']
dict_data = getAllFiles()
csv_file = "upennalerts.csv"

try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dict_data:
            writer.writerow(data)
except IOError:
    print("I/O error")

print(getAllFiles())

# readFile('UPennAlert issued at 1_18_20 3_21 AM.eml')
# print(json.dumps(parsed_eml, default=json_serial))
```
